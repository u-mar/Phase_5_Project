### Data Understanding

In this section, we provide an overview of the dataset columns and explain how each column will be utilized in our project:

**CO2 Data (c02):**
1. **Date**: The date of the CO2 measurement.
   - **Usage**: We will use this column as a timestamp to track the temporal evolution of CO2 concentrations over time.

2. **Decimal Date**: A numerical representation of the date.
   - **Usage**: This numerical representation will be used in time series analysis and potentially for visualization purposes.

3. **Average**: The average CO2 concentration in parts per million (ppm) for the given date.
   - **Usage**: This column contains the primary CO2 concentration data that we will analyze to understand trends and variations in atmospheric CO2 levels.

4. **Interpolated**: Interpolated CO2 concentration.
   - **Usage**: Interpolated data may be useful for filling in gaps or missing values in the CO2 concentration data.

5. **Trend**: The trend in CO2 concentration.
   - **Usage**: We may explore the trend component to understand the long-term behavior of CO2 concentrations.

6. **Number of Days**: The number of days associated with the measurement.
   - **Usage**: This column provides additional information about the data collection process but may not be directly used for analysis.
   
   The CO2 data consists of 727 entries and provides monthly measurements of CO2 concentration dating back to 1958.

**Temperature Data (temp):**
1. **dt**: The date of the temperature measurement.
   - **Usage**: We will use this column as a timestamp to analyze temperature patterns over time.

2. **AverageTemperature**: The average temperature for the given date and location.
   - **Usage**: This is the primary temperature data that we will analyze to understand temperature trends and variations.

3. **AverageTemperatureUncertainty**: Uncertainty associated with the temperature measurement.
   - **Usage**: Uncertainty information can be considered when assessing the reliability of temperature measurements.

4. **City**: The city where the temperature measurement was taken.
   - **Usage**: We may use city information to group temperature data by regions for regional analysis.

5. **Country**: The country where the temperature measurement was taken.
   - **Usage**: This column will be essential for regional and country-level temperature analysis.

6. **Latitude**: The latitude coordinates of the location.
   - **Usage**: Latitude information can be used for spatial analysis or mapping temperature data.

7. **Longitude**: The longitude coordinates of the location.
   - **Usage**: Longitude information complements latitude data for spatial analysis and mapping.
   
   The temperature data consists of 239,177 entries and provides monthly temperature measurements for various cities and countries dating back to 1849.

**Merged Data (merged_data_with_c02):**
1. **Date**: The date associated with CO2 measurements (from c02 data).
   - **Usage**: This date will serve as a common timestamp for merging CO2 and temperature data, enabling us to explore potential relationships between CO2 concentrations and temperature variations.

2. **Decimal Date**: A numerical representation of the date (from c02 data).
   - **Usage**: This numerical representation aligns with the "Date" column in the CO2 data and will facilitate data integration.

3. **Average**: The average CO2 concentration in parts per million (ppm) for the given date (from c02 data).
   - **Usage**: CO2 concentration data is integrated with temperature data to investigate potential links between CO2 levels and temperature patterns.

4. **Interpolated**: Interpolated CO2 concentration (from c02 data).
   - **Usage**: Interpolated data may be used for data completeness if needed in the merged dataset.

7. **AverageTemperature**: The average temperature for the given date and location (from temp data).

## Data preparation

```python
# Handle missing values ( fill with the mean)
temp['AverageTemperature'].fillna(temp['AverageTemperature'].mean())
temp['AverageTemperatureUncertainty'].fillna(temp['AverageTemperatureUncertainty'].mean())
```

```python
temp.info()
```

```python
c02['Date'] = pd.to_datetime(c02['Date'])
temp['dt'] = pd.to_datetime(temp['dt'])
```

```python
merged_data = pd.merge(c02, temp, left_on='Date', right_on='dt', how='outer')
```

```python
merged_data.info()
```

```python
merged_data.head()
```

```python
merged_data.isnull().sum()
```

```python
# Filter rows where 'Average' column is not NaN
merged_data_with_c02 = merged_data[~merged_data['Average'].isna()]

# Now, merged_data_with_c02 contains only rows with CO2 data
```

```python
# Handle missing values
merged_data_with_c02.dropna(inplace=True)

# Check for and remove duplicates
merged_data_with_c02.drop_duplicates(inplace=True)

# Check data types
print(merged_data_with_c02.dtypes)

# Check for missing values
missing_values = merged_data_with_c02.isnull().sum()
print("Missing Values:")
print(missing_values)

# Check for duplicates
duplicates = merged_data_with_c02.duplicated().sum()
print("Duplicate Rows:")
print(duplicates)
```

**Normalizing Data:** Both transformations aim to make the data more normally distributed. Many statistical and machine learning models assume that the data follows a normal distribution. Normalizing the data can improve the performance of these models.

**Box-Cox Transformation:** In the second subplot, you can see the result of the Box-Cox transformation. The 'BoxCox_AverageTemperature' column contains the transformed temperature values. ### Stationarity

Is a key assumption in many time series models, as it ensures that the statistical properties of the time series data do not change over time. To check for stationarity, we can use statistical tests such as the Augmented Dickey-Fuller (ADF) test.

```python
# Perform the Augmented Dickey-Fuller (ADF) test on the original data
adf_result = adfuller(temp['AverageTemperature'])
print('ADF Statistic:', adf_result[0])
print('p-value:', adf_result[1])
print('Critical Values:', adf_result[4])

# Interpret the ADF test results
if adf_result[1] <= 0.05:
    print('The time series is stationary (reject the null hypothesis)')
else:
    print('The time series is not stationary (fail to reject the null hypothesis)')
```

The Augmented Dickey-Fuller (ADF) test results indicate that our time series data is stationary since the p-value is less than the significance level of 0.05, and the ADF statistic is highly negative. This suggests that we can proceed with time series modeling without the need for differencing to achieve stationarity.

### Extracting Seasonal Features

Seasonal features can improve the accuracy of predictive models. By including information about seasons, we can make more informed predictions about future temperature trends, taking into account seasonal fluctuations.

```python
# Extract the month and season from the 'dt' column
temp['Month'] = temp.index.month

# Define a function to map months to seasons
def get_season(month):
    if 3 <= month <= 5:
        return 'Spring'
    elif 6 <= month <= 8:
        return 'Summer'
    elif 9 <= month <= 11:
        return 'Fall'
    else:
        return 'Winter'

# Apply the get_season function to create a 'Season' column
temp['Season'] = temp['Month'].apply(get_season)

# Create binary columns for each season using one-hot encoding
temp = pd.get_dummies(temp, columns=['Season'], prefix='Season', drop_first=True)

# Create one-hot encoding for the month of the year
temp = pd.get_dummies(temp, columns=['Month'], prefix='Month', drop_first=True)

# Display the updated DataFrame with seasonal features
print(temp.head())
```

These features will help our models capture seasonal variations in the data.

### Cyclical Features for Month and Day

Cyclical features for "Month" and "Day" are created to capture periodic patterns in time data, such as monthly and daily cycles. These features are helpful in time series analysis and predictive modeling.

```python
# Cyclical Features for Month and Day
def create_cyclical_features(df, column, period):
    df[f'{column}_sin'] = np.sin(2 * np.pi * df[column] / period)
    df[f'{column}_cos'] = np.cos(2 * np.pi * df[column] / period)

# Extract the month and season from the 'Date' column
temp['Month'] = temp.index.month

# Create cyclical features for Month (period = 12 for annual cycle)
create_cyclical_features(temp, 'Month', period=12)

# Extract the day of the month from the 'Date' column
temp['Day'] = temp.index.day

# Create cyclical features for Day (period = 31 for monthly cycle)
create_cyclical_features(temp, 'Day', period=31)

# Display the updated DataFrame with all features
print(temp.head())
```

For "Month," sine and cosine transformations are applied to represent its cyclical nature with a period of 12 (for the annual cycle).
For "Day," sine and cosine transformations are applied to represent its cyclical nature with a period of 31 (for the monthly cycle).
The resulting features, "Month_sin," "Month_cos," "Day_sin," and "Day_cos," encode the cyclicality of months and days, making it easier for models to capture seasonal variations in the data. These features encode the cyclicality of months and days, allowing our models to capture periodic patterns.
