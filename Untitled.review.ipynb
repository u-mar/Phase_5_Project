**CO2 Data (c02):**
1. **Date**: The date of the CO2 measurement.

   - **Usage**: Interpolated data may be useful for filling in gaps or missing values in the CO2 concentration data.

We will now be looking froward to have both of our datasets merged so as to produce a uniform analysis

```python
# Putting them in dataframe
c02['Date'] = pd.to_datetime(c02['Date'])
temp['dt'] = pd.to_datetime(temp['dt'])
```

### Exploratory Data Analysis

```python
# Create a line plot of CO2 concentration over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=merged_data_with_c02, x='Date', y='CO2_Average', label='CO2 Concentration (ppm)')
plt.title('CO2 Concentration Over Time')
plt.xlabel('Date')
plt.ylabel('CO2 Concentration (ppm)')
plt.xticks(rotation=45)
plt.show()
```

### Handling zero or negative values in the 'AverageTemperature' column is essential before applying transformations

```python
# Check for zero or negative values in the 'AverageTemperature' column
negative_temp_indices = temp[temp['AverageTemperature'] <= 0].index

if not negative_temp_indices.empty:
    # Handle zero or negative values by adding a small positive constant
    small_positive_constant = 1
    temp.loc[negative_temp_indices, 'AverageTemperature'] = small_positive_constant
```

We'll perform both the logarithmic transformation and the Box-Cox transformation on the "AverageTemperature" column to stabilize its variance. This can help improve the performance of statistical models that assume constant variance.

```python
# Logarithmic transformation
temp['Log_AverageTemperature'] = np.log(temp['AverageTemperature'] + 1)  # Adding 1 to avoid issues with zero or negative values

# Box-Cox transformation
from scipy import stats
lmbda = 0.15  # Example lambda value
temp['BoxCox_AverageTemperature'] = stats.boxcox(temp['AverageTemperature'] + 1, lmbda=lmbda)

# Plot the transformed data for visualization
plt.figure(figsize=(12, 6))

# Logarithmic transformation
plt.subplot(121)
plt.plot(temp.index, temp['Log_AverageTemperature'], label='Log Transformation', color='blue')
plt.title('Logarithmic Transformation of Temperature')
plt.xlabel('Date')
plt.ylabel('Log(Average Temperature)')
plt.grid(True)

# Box-Cox transformation
plt.subplot(122)
plt.plot(temp.index, temp['BoxCox_AverageTemperature'], label='Box-Cox Transformation', color='green')
plt.title('Box-Cox Transformation of Temperature')
plt.xlabel('Date')
plt.ylabel('Box-Cox(Average Temperature)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

**Why perfom transformations?**

**Stabilizing Variance:** These transformations are used to mitigate issues related to heteroscedasticity, where the variance of the data is not constant across the time series. By stabilizing variance, we make it easier for time series models to capture patterns and relationships in the data.

**Normalizing Data:** Both transformations aim to make the data more normally distributed. Many statistical and machine learning models assume that the data follows a normal distribution. Normalizing the data can improve the performance of these models.

**Results:**

**Logarithmic Transformation:** In the first subplot, you can see the result of the logarithmic transformation. The 'Log_AverageTemperature' shows temperature values that have been transformed using the natural logarithm. This transformation compresses higher values and expands lower values, making extreme temperature values less extreme. It can help make temperature data variations more consistent.

**Box-Cox Transformation:** In the second subplot, you can see the result of the Box-Cox transformation. The 'BoxCox_AverageTemperature' column contains the transformed temperature values. This transformation is more flexible and can adapt to the data's distribution. In this case, it reduces the skewness and stabilizes variance.

**Stationarity** is a key assumption in many time series models, as it ensures that the statistical properties of the time series data do not change over time. To check for stationarity, we can use statistical tests such as the Augmented Dickey-Fuller (ADF) test.

```python
# Perform the Augmented Dickey-Fuller (ADF) test on the original data
adf_result = adfuller(temp['AverageTemperature'])
print('ADF Statistic:', adf_result[0])
print('p-value:', adf_result[1])
print('Critical Values:', adf_result[4])

# Interpret the ADF test results
if adf_result[1] <= 0.05:
    print('The time series is stationary (reject the null hypothesis)')
else:
    print('The time series is not stationary (fail to reject the null hypothesis)')
```

The Augmented Dickey-Fuller (ADF) test results indicate that our time series data is stationary since the p-value is less than the significance level of 0.05, and the ADF statistic is highly negative. This suggests that we can proceed with time series modeling without the need for differencing to achieve stationarity.
